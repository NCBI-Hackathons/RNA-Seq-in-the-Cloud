```python
def gen_metadata_colors(series):
    unique = series.unique()
    num_colors = len(unique)
    pal = sns.color_palette("Set2",num_colors)
    hex_codes = pal.as_hex()
    _d = dict(zip(unique, hex_codes))
    colors = series.map(_d)
    return colors

# gen_metadata_colors(metadata_test['Sex'])
```

# PCA

```python
# Run PCA
def plot_pca(data_df, metadata_df,metadata_OI):
    # need to separate out the PCA components by the metadata coloring of choice so that the legend is linked correctly
    
    pca=PCA(n_components=3)
    pca.fit(data_df)
#     print(pca.components_)
    # Get variance explained
    var_explained = ['PC'+str((i+1))+'('+str(round(e*100, 1))+'% var. explained)' for i, e in enumerate(pca.explained_variance_ratio_)]
    
    # Plot PCA
    trace = go.Scatter3d(x=pca.components_[0],
                y=pca.components_[1],
                z=pca.components_[2],
                mode='markers',
                hoverinfo='text',
                text=['<br>'.join(['{key}: {value}'.format(**locals()) for key, value in rowData.items()]) for index, rowData in metadata_df.loc[data_df.columns].rename_axis('Signature ID').reset_index().iterrows()],
                marker=dict(color=gen_metadata_colors(metadata_df[metadata_OI]), size=15, opacity=0.8))
    #             marker={'size': 15, 'showscale': False, 'colorscale': 'Jet', 'color': metadata_dataframe.loc[signature_dataframe.columns]['pert_time']})
    data = [trace]
    layout = go.Layout(title='PCA Analysis | Scatter Plot<br><span style="font-style: italic;">Colored by '+metadata_OI+' </span>', hovermode='closest', margin=go.Margin(l=0,r=0,b=0,t=50), width=900, scene=dict(xaxis=dict(title=var_explained[0]), yaxis=dict(title=var_explained[1]),zaxis=dict(title=var_explained[2])), showlegend=True)
    fig = go.Figure(data=data, layout=layout)
#     fig.legend()
    iplot(fig)
```

```python
plot_pca(data_df_test, metadata_test,'Genotype')
```
